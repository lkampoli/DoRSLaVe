# Overview

These are the lecture slides for the **"Design of Rockets and Space Launch
Vehicles"** course from The University of Sydney. They can be used to see and
follow which topics are covered in the course. The slides are not intended to
be used to learn the design of rockets and space launch vehicles from scratch.
For that, you need to attend the lectures and complete the assignments and
assessment sessions.
